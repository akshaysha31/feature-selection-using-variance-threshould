```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
df=pd.read_csv("train(1).csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.drop(['target',],axis=1)

y=df['target']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
x_train=x_train.drop("ID_code",axis=1)
```

### apply the variance threshold

```python
var_the=VarianceThreshold(threshold=0)
var_the.fit(x_train)
```

```python
var_the.get_support()
```

 let find the non constant features

```python
len(var_the.get_support())
```

```python
len(x_train.columns[var_the.get_support()])
```

```python
constant_column=[column for column in x_train.columns if column not in x_train.columns[var_the.get_support()]]
```

```python
print(len(constant_column))
```

```python
for column in constant_column:
    print(column)
```

```python
x_train.drop(constant_column,axis=1)
```
